#### Installing Samba:

To install samba on the raspberry pi the following command was used:

* pi@combspi4:~ $ sudo apt-get install samba samba-common-bin

The installation of samba was verified and located using the following commands:

* pi@combspi4:~ $ sudo /etc/init.d/smbd status
* pi@combspi4:~ $ whereis samba

The location of the samba is: samba: /usr/sbin/samba /usr/lib/arm-linux-gnueabihf/samba /etc/samba /usr/share/samba /usr/share/man/man8/samba.8.gz /usr/share/man/man7/samba.7.gz

After samba was verified, I changed the password on my samba using the command below. This allows for me to gain access the raspberry pi as a shared server.

* sudo smbpasswd -a

I then signed into the raspberry pi established server from my mac. This is where I ran into my largest issue. I was unable to share files between my mac and the raspberry pi. The solution was adjust the configuration of the same to make sure that I could read and write to the shared folder that I wanted. 

This is the command use to access the samba configuration: 

* pi@combspi4:~ $ sudo nano /etc/samba/smb.conf

Below are the modifications I made that allowed me to file transfer. I am not sure if this is the most efficient way, but after reading several articles and trial and error, this was the solution that allowed me to transfer files succesfully. I changed [home] to read only = no and add the below code:

[eep522]

   path = /home/pi/eep522
   
   read only = no
   
   browsable = yes
   
I also uncommented this section:

[netlogon] 

   comment = Network Logon Service
   
   path = /home/samba/netlogon
   
   guest ok = yes
   
   read only = no

From here I was able to drag and drop files in the shared folder and watch them appear in the GUI files on the raspberry pi monitor.
